```python
# Name input files and open file handles to them
file_A_in = 'algaeAnnotate.tab'
file_B_in = 'coralAnnotate.tab'
fhinA  = open( file_A_in  , "r"  )
fhinB  = open( file_B_in  , "r"  )

# create two empty dictionaries 
dict_A_all_lines = {}
dict_B_all_lines = {}

# The following loop iterates over each line in 'algaeAnnotate.tab'
# and for each line first splits the line on tabs into an array of 
# the cells of the line; then stores as separate variables
# the contig, evalue, and bitscore; next adds to the dict_A_all_lines
# the contig a a key and the value associated is the 
# eval and a tab and a bitscore.

for line in fhinA:
	row_A_i_cells = line.split( "\t" )
	contig_A_i = row_A_i_cells[0]
	eval       = row_A_i_cells[-2]
	bitscore   = row_A_i_cells[-1]
	dict_A_all_lines[contig_A_i] = eval + "\t" + bitscore
    
# Does same thing with 'coralAnnotate.tab' into dict_B_all_lines
# that previous loop did to 'algaeAnnotate.tab' into dict_A_all_lines.
for line in fhinB:
	row_B_i_cells = line.split( "\t" )
	contig_B_i = row_B_i_cells[0]
	eval       = row_B_i_cells[-2]
	bitscore   = row_B_i_cells[-1]
	dict_B_all_lines[contig_B_i] = eval + "\t" + bitscore
    
fhinA.close()
fhinB.close()
```

```python
# Now put into dict_A_contigs_evalsPlusTabPlusBitscore
# the data from dict_A_all_lines s.t. that contig also
# existed as a key in dict_A_all_lines
# Then do same with other file's subset. ...
# Now put into dict_B_contigs_evalsPlusTabPlusBitscore
# the data from dict_B_all_lines s.t. that contig also
# existed as a key in dict_B_all_lines
#
# When done with both of following loops, the size of the
# 2 dictionaries shall be 
# equal (and the keys shall be the same, but corresponding
# values shall be different).

dict_A_contigs_evalsPlusTabPlusBitscore = {}
dict_B_contigs_evalsPlusTabPlusBitscore = {}
for contig in dict_A_all_lines.keys():
	if contig in dict_B_all_lines:
		dict_A_contigs_evalsPlusTabPlusBitscore[contig] = dict_A_all_lines[contig]	
for contig in dict_B_all_lines.keys():
	if contig in dict_A_all_lines:
		dict_B_contigs_evalsPlusTabPlusBitscore[contig] = dict_B_all_lines[contig]

# Name output files and open file handles to them.
file_A_out = 'algae.tab'
file_B_out = 'coral.tab'
fhoutA = open( file_A_out , "w+" )
fhoutB = open( file_B_out , "w+" )
```

```python
# sort the keys so that output will be easier to read because the contigs
# of each output file will be alpha-sorted.

for contig in sorted(dict_A_contigs_evalsPlusTabPlusBitscore.keys()):
# The evals and bitscores are tab-separated as one combined string 
# of info which is the value in each list corresponding to some contig key.
# the splits below pull the evals and bitscores out for comparison and
# determination of correct output file.

	eval_from_A     = float(dict_A_contigs_evalsPlusTabPlusBitscore[contig].split( "\t" )[0])
	eval_from_B     = float(dict_B_contigs_evalsPlusTabPlusBitscore[contig].split( "\t" )[0])
	bitscore_from_A = float(dict_A_contigs_evalsPlusTabPlusBitscore[contig].split( "\t" )[1])
	bitscore_from_B = float(dict_B_contigs_evalsPlusTabPlusBitscore[contig].split( "\t" )[1])
	
# if algae's eval is < coral's eval, then save to algae file the contig and 
# corresponding eval and bitscore for algae.
	if eval_from_A < eval_from_B:
		fhoutA.write( contig + "\t" + dict_A_all_lines[contig] )
# if coral's eval is < algae's eval, then save to coral file the contig and 
# corresponding eval and bitscore for coral.
	elif eval_from_B < eval_from_A:
		fhoutB.write( contig + "\t" + dict_B_all_lines[contig]  )
# if evals are equal for the algae and the coral, then determine which
# bitscore is higher.
	elif eval_from_A == eval_from_B:
# if algae bitscore > coral bitscore, then write to algae file.
		if bitscore_from_A > bitscore_from_B:
			fhoutA.write( contig + "\t" + dict_A_all_lines[contig] )
# if coral bitscore > algae bitscore, then write to coral file.
		elif bitscore_from_B > bitscore_from_A:
			fhoutB.write( contig + "\t" + dict_B_all_lines[contig] )
	
fhoutA.close()
fhoutB.close()
```
